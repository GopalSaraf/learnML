```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
dataset = pd.read_csv("data/social_ads.csv")
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
from learnML.preprocessing import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
from learnML.preprocessing import ZScoreNormalization

# Normalize the data
x_scalar = ZScoreNormalization(X_train)
x_scalar.fit_transform()
```

```python
from learnML.classification import LogisticRegression

# Train the model
model = LogisticRegression(learning_rate=0.001, n_iterations=100000, x_scalar=x_scalar)
model.fit(X_train, y_train)
```

```python
print("Weights:", model.get_weights())
print("Intercept:", model.get_intercept())
```

```python
y_pred = model.predict(X_test)
np.concatenate((y_pred.reshape(-1, 1), y_test.reshape(-1, 1)), axis=1)
```

```python
from learnML.utils import (
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
)

confusion_matrix(y_pred, y_test)
```

```python
print("Accuracy:", accuracy_score(y_pred, y_test))
print("Precision:", precision_score(y_pred, y_test))
print("Recall:", recall_score(y_pred, y_test))
```

```python
plt.plot(model.get_cost_history())
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.title("Cost per iteration")
plt.show()
```
